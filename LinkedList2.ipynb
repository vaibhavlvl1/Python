```python
# Basic Input Output

class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        

        
def LinkedListInput():
    arr = [int(x) for x in input().split()]
    head = None
    
    for currData in arr:
        if currData == -1:
            break
        newNode = Node(currData)
        if head is None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
    
    return head

def length(head):
    count = 0
    temp = head                # avoiding head to become none # it doesnt matter anyway
    while temp is not None:
        count +=1
        temp = temp.next
    return count

def printLL(head):
    temp = head
    while temp is not None:
        print(temp.data,end="-->")
        temp = temp.next
    print("None")
def midpoint(head):
    slow = head
    fast = head
    
    while fast.next is not None and fast.next.next is not None:
        slow  = slow.next
        fast = fast.next.next
        
    return slow.data
```

```python
# reverse using recursion

def reverseR(head):
    if head is None or head.next is None:
        return head
    smallhead= reverseR(head.next)
    
    head.next.next = head
    head.next = None
    return smallhead

head = LinkedListInput()
newHead = reverse(head)
printLL(newHead)
```

```python
# reverse using iteration

def reverseI(head):
    curr = head
    prev = None
    
    while curr is not None:
        nexti = curr.next
        curr.next = prev
        prev = curr
        curr = nexti
        
    return prev

head = LinkedListInput()
printLL(reverse(head))
```

```python
# merge two sorted Linked List

def merge(head1,head2):
    if head1 is None:
        return head2
    elif head2 is None:
        return head1
    head = Node(None)
    tail = head
    
    while head1 is not None and head2 is not None:
        if head1.data< head2.data:
            tail.next = head1
            tail = tail.next
            head1 = head1.next
        else: #head1.data>head2.data:
            tail.next = head2
            tail = tail.next
            head2 = head2.next
            
    while head1 is not None:
        tail.next = head1
        tail = tail.next
        head1 = head1.next
    while head2 is not None:
        tail.next = head2
        tail = tail.next
        head2 = head2.next
        
    return head.next


```

```python
# code merge sort for Linked Lists
# the merge function above will be used 

# needs a midpoint function which return mid node

def midpoint2(head):
    fast = head
    slow = head
    while fast.next is not None and fast.next.next is not None:
        slow = slow.next
        fast = fast.next.next
    return slow

def merge_sort(head):
    if head is None or head.next is None:
        return head
    
    mid = midpoint2(head)
    head1 = mid.next
    mid.next = None
    
    newHead1 = merge_sort(head)
    newHead2 = merge_sort(head1)
    
    return merge(newHead1,newHead2)
    

head = LinkedListInput()
newHead = merge_sort(head)
printLL(newHead)
```

```python
# Find a node in linked List Recursive

def find_node(head,data,index):
    if head is None:
        return -1
    if head.data == data:
        return index
    return find_node(head.next,data,index+1)


head = LinkedListInput()
data = int(input())
find_node(head,data,0)
```

```python
# skipMdeleteN
# recursion 
# ridded the need for use of prev by using count == 1 this makes curr pointer to point
# to the skip and delete end ele so that the their next is dele and skip part respectively

def skipMdeleteN(head,m,n):
    if head is None:
        return head
    if m ==0:
        return
    if n==0:
        return head
    
    c = 0
    curr = head
    prev = None
    
    while curr is not None:
        if c==m:
            break
        c = c+1
        prev = curr
        curr = curr.next
    
    d = 0
    while curr  is not None:
        if d==n:
            break
        d = d+1
        curr = curr.next
    
    prev.next = skipMdeleteN(curr,m,n)
    
    return head

head = LinkedListInput()
m = int(input())
n = int(input())
newhead = skipMdeleteN(head,m,n)
printLL(newhead)
```

```python
# kreverse
# all cases passed

def reverse3(head):
    if head is None:
        return
    curr = head
    prev = None
    while curr is not None:
        save = curr.next
        curr.next = prev
        prev = curr
        curr = save
    return prev,head


def kreverse(head,k):
    if k==0 or k ==1:
        return head
    if head is None:
        return head
    c = 0
    curr = head
    while curr is not None:
        if c==k:
            break
        c = c+1
        prev = curr
        curr = curr.next
    prev.next = None
        
    newhead,newtail = reverse3(head)
    
    newtail.next = kreverse(curr,k)
    
    return newhead




head = LinkedListInput()
newhead= kreverse(head,1)
printLL(newhead)
```
