```python
# Binary Pattern

"""
n = 4
1111
000
11
1
"""

n = int(input())

for i in range(n):
    
    for j in range(n-i):
        
        if i%2==0:
            print("1",end="")
        else:
            print("0",end="")
    print()


```

```python
# using while loop

n = int(input())

i = 0
while i<=n:
    j = n-i
    while j!=0:
        if i%2==0:
            print("1",end="")
        else:
            print("0",end="")
        j = j-1
    print()
    i = i+1
```

```python
##Print Number Pyramid

"""
n = 6
123456
 23456
  3456
   456
    56
     6
    56
   456
  3456
 23456
123456 
"""


def numPyramid(n):
    # First Half
    for i in range(1,n+1):
        # first half spaces
        for j in range(i-1):
            print(".",end="")
        # first half numbers
        m = i
        for j in range(n-i,-1,-1):
            print(m,end="")
            m = m+1
        print()
    # second half
    n1 = n-1
    for i in range(n1):
        # spaces
        for j in range(n1-i-1):
            print(".",end = "")
        # numbers
        x = n-i-1
        for j in range(n-x+1):
            print(x,end="")
            x = x+1
            
            
        print()
        
    
    
n =int(input())
numPyramid(n)
```

```python
# Diamond of stars
"""
n = 5
  *
 ***
*****
 ***
  *
"""


def printStars(n):
    N1 = (n+1)//2
    N2 = n-N1
    # First Half
    for i in range(N1):
        # space  equation is (N-i)
        for j in range(N1-i-1):
            print(" ",end="")
        
        # star equation is (2i-1)
        for j in range(2*i+1):
            print("*",end = "")
        print()
        
        
        
    # second half 
    for i in range(1,N2+1):
        # space equation is i
        
        for j in range(i):
            print(" ",end="")
            
        # star print 
            
        for j in range(2*N2 - 2*i +1):
            print("*",end="")
            
        print()
        
        
n = int(input())
printStars(n)
            
```

```python
# print the pattern 

"""
n = 5
1 2 3 4 5
11 12 13 14 15
21 22 23 24 25
16 17 18 19 20
6 7 8 9 10
"""

n = int(input())

N1 = (n+1)//2
N2 = n-N1

for i in range(N1):
    for j in range(1,n+1):
        print((n*2*i)+j,end=" ")
    print()
        
for i in range(N2,0,-1):
    for j in range(1,n+1):
        print(n*(2*i-1)+j,end=" ")
    print()
    
```
