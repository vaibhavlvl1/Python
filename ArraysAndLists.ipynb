```python
# swap alternate
# time complexity = O(n)

def printArr(arr):
    for ele in arr:
        print(ele,end = " ")
    print()

def swap(arr):
    for i in range(0,len(arr)-1,2):
        arr[i],arr[i+1]=arr[i+1],arr[i]
        

        
#driver
t = int(input())
while t!=0:
    n = int(input())
    arr = [int(x) for x in input().split()]
    swap(arr)
    printArr(arr)
    t =t-1
        
```

```python
# Find Unique
# using dictionary
# TC = O(n)

def findUnique(arr):
    d = {}
    for ele in arr:
        d[ele] = d.get(ele,0)+1
        
    for ele in arr:
        if d[ele] ==1:
            return ele
        
t = int(input())

while t!=0:
    n = int(input())
    arr =  [int(x) for x in input().split()]
    print(findUnique(arr))
    t = t-1
```

```python
# find duplicate
# TC = O(n)

def findDupli(arr):
    d = {}
    for i in arr:
        d[i] = d.get(i,0)+1
        
    for ele in arr:
        if d[ele] >1:
            return ele
        
t = int(input())
while t!=0:
    n = int(input())
    arr = [int(x) for x in input().split()]
    print(findDupli(arr))
    t = t-1
    
```

```python
#intersection of two arrays
# TC = O(n1+n2)

def printArr(arr):
    for ele in arr:
        print(ele,end=" ")
    print()
    
def intersection(arr1,arr2):
    d = {}
    res = []
    for ele in arr2:
        d[ele] = d.get(ele,0)+1
        
    for ele in arr1:
        if d.get(ele,0) !=0:
            res.append(ele)
            d[ele]-=1
    return res
            

    
t = int(input())
while t!=0:
    n1 = int(input())
    arr1 = [int(x) for x in input().split()]
    n2 = int(input())
    arr2 = [int(x) for x in input().split()]
    res = intersection(arr1,arr2)
    printArr(res)
    t = t-1

            
    
```

```python
# pair sum
# bruteForce method
# TC = O(n*m)


def pairSum(arr,target):
    n = len(arr)
    pairCount = 0
    for i in range(n):
        for j in range(i+1,n):
            if arr[i] +arr[j] ==target:
                #print(arr[i],arr[j])
                pairCount +=1
                
    return pairCount


t = int(input())
while t!=0:
    n = int(input())
    if n==0:
        print("0")
        break
    arr = [int(x) for x in input().split()]
    target = int(input())
    print(pairSum(arr,target))
    t = t-1

    
```

```python
# Dictionary Method
# TC = O(n)

def pairSum(arr,target):
    count = 0
    d = {}
    
    for i in arr:
        if target - i in d:
            count = count + d[target-i]
        
        d[i]=d.get(i,0)+1
    
            
        
            
    return count

t = int(input())
while t!=0:
    n = int(input())
    if n==0:
        print("0")
        break
    arr = [int(x) for x in input().split()]
    target = int(input())
    print(pairSum(arr,target))
    t = t-1

        
```

```python
## Triplet Sum
# time complexity O(n**3)

def tripletSum(arr,t):
    count = 0
    for i in range(n):
        for j in range(i+1,n):
            for k in range(j+1,n):
                if arr[i]+arr[j]+arr[k]==t:
                    count = count+1
                    
    return count


t = int(input())

while t!=0:
    n = int(input())
    if n==0:
        print("0")
        break
    arr = [int(x) for x in input().split()]
    target = int(input())
    print(tripletSum(arr,target))
    t = t-1
```

```python
# using dictionary method
# Time Complexity  = O(n*m)

def tripletSum(arr,target):
    count = 0
    for i in range(len(arr)):
        count = count + pairSum(arr[i+1:],target-arr[i])
        
    return count



t = int(input())

while t!=0:
    n = int(input())
    if n==0:
        print("0")
        break
    arr = [int(x) for x in input().split()]
    target = int(input())
    print(tripletSum(arr,target))
    t = t-1
```

```python
# sort 0 and 1


def sort01(arr):
    n = len(arr)
    left,right = 0, n-1
    
    while left<=right:
        if arr[left] ==1 and arr[right]==0:
            arr[left],arr[right] = arr[right],arr[left]
            left +=1
            right -=1
            
        if arr[left] ==0 and arr[right]==1:
            left = left+1
            right = right +1
        if arr[left] ==1 and arr[right] ==1:
            right-=1
        if arr[left] ==0 and arr[right] ==0:
            left = left+1
            
arr = [1,1,1,0,0,0,1,0,1,0]
sort01(arr)
print(arr)
```
