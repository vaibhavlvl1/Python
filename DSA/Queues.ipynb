```python
# Implementation of Queue using Array

class queue:
    def __init__(self):
        self.arr = []
        self.count = 0
        self.queueStart = 0
    
    def isEmpty(self):
        return self.size()==0
        
    def size(self):
        return self.count
    def front(self):
        if self.size()==0:
            return -1
        return self.arr[self.queueStart]
        
    def enqueue(self,data):
        self.arr.append(data)
        self.count +=1
    def dequeue(self):
        if self.size()==0:
            return -1
        ele = self.arr[self.queueStart]
        self.queueStart+=1
        self.count -=1
        return ele
        
    def printQ(self):
        print(self.arr[self.queueStart:])
        
```

```python
# stack using two queues

import queue


class stacks:
    def __init__(self):
        self.q1 = queue.Queue()
        self.q2 = queue.Queue()
        
    def size(self):
        return self.q1.qsize()
    def isEmpty(self):
        return self.size()==0
    
    def push(self,data):
        self.q1.put(data)
        
    def pop(self):
        if self.size()==0:
            return -1
        while self.size()>1:
            self.q2.put(self.q1.get())
            
        x = self.q1.get()
        
        temp = self.q2
        self.q2 = self.q1
        self.q1 = temp
        return x
    
    def top(self):
        if self.q1.qsize()==0:
            return -1
        while self.size()>1:
            self.q2.put(self.q1.get())
            
        x = self.q1.get()
        self.q2.put(x)
        
        temp = self.q1
        self.q1 = self.q2
        self.q2 = temp
        
        
        
        return x
        
```
