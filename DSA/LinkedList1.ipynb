```python
# Basic linked List functions 
# will be used throughout the file
# implementation, input,length, printing



class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        

        
def LinkedListInput():
    arr = [int(x) for x in input().split()]
    head = None
    
    for currData in arr:
        if currData == -1:
            break
        newNode = Node(currData)
        if head is None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
    
    return head

def length(head):
    count = 0
    temp = head                # avoiding head to become none # it doesnt matter anyway
    while temp is not None:
        count +=1
        temp = temp.next
    return count

def printLL(head):
    temp = head
    while temp is not None:
        print(temp.data,end="-->")
        temp = temp.next
    print("None")
def midpoint(head):
    slow = head
    fast = head
    
    while fast.next is not None and fast.next.next is not None:
        slow  = slow.next
        fast = fast.next.next
        
    return slow.data
```

```python
# print ith Node

def printIthNode(head,i):
    c = 0
    temp = head
    while temp is not None:
        if c==i:
            print(temp.data)
            return
        c = c+1
        temp = temp.next
    return

head = LinkedListInput()
i = int(input())
printIthNode(head,i)

```

```python
# delete a node 


def deleteNode(head, pos) :
    if pos == 0:
        if head is None:             # if pos is 0 and ll is None
            return
        else:
            head = head.next         # if only 1 ele in list
            return head
    c = 0
    temp = head
    prev = None
    
    while temp is not None:
        if c == pos:
            prev.next = temp.next
            temp.next = None
        
        c = c+1
        prev = temp
        temp = temp.next
    return head
        

# Main
t = int(input())
while t!=0:
    head = LinkedListInput()
    i = int(input())
    newHead = deleteNode(head,i)
    printLL(newHead)
    t = t-1
    

            
    
```

```python
# length of ll Recursive

def lengthR(head):
    if head is None:
        return 0
    
    return 1+lengthR(head.next)

head = LinkedListInput()
print(lengthR(head))
```

```python
# delete Node Recursive

def deleteR(head,pos):
    if pos <0:
        return head
    if head is None:
        return None
    
    if pos == 0:
        return head.next
    head.next = deleteR(head.next,pos-1)
    return head
    
head = LinkedListInput()
newHead = deleteR(head,int(input()))
printLL(newHead)
    
```

```python
# Find a Node in Linked List

def findNode(head,nodeData):
    pos = 0
    temp = head
    while temp is not None:
        if temp.data == nodeData:
            return pos
        pos = pos+1
        temp = temp.next
    return -1
        
        

t = int(input())
while t!=0:
    head = LinkedListInput()
    toSearch = int(input())
    print(findNode(head,toSearch))
    t = t-1


    
```

```python
# AppendLastNtoFirst


def append(head,n):
    if head is None:
        return None
    if n==0 or n == length(head):
        return head
    k = length(head)  # this is a function
    c = 0
    temp = head
    while temp is not None:
        if c == (k-n)-1:
            newHead = temp.next
            temp.next = None
            break
        c = c+1
        temp = temp.next
    temp2 = newHead
    while temp2 is not None:
        tail = temp2
        temp2 = temp2.next
    tail.next = head
    return newHead

head = LinkedListInput()
n = int(input())
newHead = append(head,n)
printLL(newHead)
```

```python
# Eliminate Duplicates from LL 


def eliminate(head):
    temp = head
    
    while temp.next is not None:
        if temp.data == temp.next.data:
            temp.next = temp.next.next
        else:
            temp = temp.next
    return head


head = LinkedListInput()
newhead = eliminate(head)
printLL(head)
```

```python
# reverse Linked List Recursive
# just printing and no changing pointers

def reverse(head):
    if head is None:
        return
    reverse(head.next)
    print(head.data,end=" ")
    

    
    
head = LinkedListInput()
reverse(head)
```

```python
# reverse Linked List

def reverse2(head):
    prev = None
    curr = head
    
    while curr is not None:
        nexti = curr.next
        curr.next = prev
        prev = curr
        curr = nexti
    newhead = prev
    return newhead

        
head = LinkedListInput()
newHead = reverse2(head)
printLL(newHead)
    
```

```python
# Palindrome Linked List


def midpoint2(head):
    slow = head
    fast = head
    
    while fast.next is not None and fast.next.next is not None:
        slow  = slow.next
        fast = fast.next.next
        
    return slow
def palindrome(head):
    mid = midpoint2(head)
    head2 = mid.next
    mid.next = None
    
    revHead2 = reverse2(head2)
    
    while head is not None and revHead2 is not None:
        if head.data != revHead2.data:
            return False
        
        head = head.next
        revHead2 = revHead2.next
        
    
    return True
        

head = LinkedListInput()
palindrome(head)
    
```
