```python
#Implementation of Stacks Using List
class stack:
    
    def __init__(self):
        self.data = []
        self.count = 0
        
        
    def push(self,item):
        self.data.append(item)
        self.count+=1
        
    def pop(self):
        if self.isEmpty()==True:
            return "Already Empty"
        else:
            x = self.data.pop()
            self.count-=1
            return x
    def top(self):
        if self.count==0:
            return "Already Empty"
        return self.data[self.count-1]
    def isEmpty(self):
        return self.count ==0
    def size(self):
        return self.count
    def printStack(self):
        print(*self.data)
        

```
