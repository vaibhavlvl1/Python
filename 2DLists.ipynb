```python
# row wise sum

def rowWise(n,m,arr):
    for i in range(n):
        rowsum = 0
        for j in range(m):
            rowsum += arr[i][j]
        
        print(rowsum,end=" ")
    
    

    
    
def takeInput():
    N,M = [int(x) for x in input().split()]
    if N==0:
        return 0,0,list()
    arr = [[int(x)for x in input().split()]for i in range(N)]
    
    return N,M,arr
    
    
t = int(input())

while t!=0:
    N,M ,arr = takeInput()
    
    rowWise(N,M,arr)
    print()           # this is imp if test cases more than 1
    t = t-1
```

```python
# wave printing an array

def wavePrint(n,m,li):
    for j in range(m):
        for i in range(n):
            if j%2==0:
                print(li[i][j],end=" ")
                
        for i in range(n-1,-1,-1):
            if j%2!=0:
                print(li[i][j],end=" ")

t = int(input())

while(t!=0):
    n,m = [int(ele) for ele in input().split()]
    a = [[int(x) for x in input().split()]for i in range(n)]
    wavePrint(n,m,a)
    print()
    t= t-1
```

```python
# spiral printing

def spiralPrinting(m,n,arr):
    row = 0
    col = 0
    
    while row < m and col <n :
        
        for j in range(col,n):
            print(arr[row][j],end=" ")
        row +=1
        
        for i in range(row,m):
            print(arr[i][n-1],end=" ")
        n -=1
        
        if row<m:
            for i in range(n-1,col-1,-1):
                print(arr[m-1][i],end=" ")
            m = m-1
        if col<n:
            for i in range(m-1,row-1,-1):
                print(arr[i][col],end=" ")
            col = col+1
        
        

                
                
                

def takeInput():
    n,m = [int(x) for x in input().split()]
    if n==0:
        return 0,0,list()
    arr = [[int(x) for x in input().split()]for i in range(n)]
    return n,m,arr

t = int(input())

while t!=0:
    n,m,arr = takeInput()
    if n==0:
        break
    spiralPrinting(n,m,arr)
    print()
    t = t-1
```

```python
# method 2

def spiralPrint(n,m,arr):
    
    l = 0
    t = 0
    r = m
    d = n
    
    while l<r and t<d:
        
        for i in range(l,r):
            print(arr[t][i],end= ' ')
        
        for j in range(t+1,d-1):
            print(arr[j][r-1],end=" ")
            
        
        
        if d!=t+1:
            for i in range(r-1,l-1,-1):
                print(arr[d-1][i],end=" ")

        if l!=r-1:
            for j in range(d-2,t,-1):
                print(arr[j][l],end=" ")
        

        t +=1
        d -=1
        l +=1
        r -=1
arr = [[1,2,3],[4,5,6],[7,8,9]]
n,m = 3,3

spiralPrint(n,m,arr)
```
